```python

import numpy as np
import matplotlib.pyplot as plt
x=np.array([[1,1],[1,2],[1,3],[1,4]])
y=np.array([1,3,5,8])

Xt=np.transpose(x);
beta=np.linalg.inv(Xt@x)@Xt@y
print("A^:",beta)
print("Slope:",beta[1])
print("Intercept:",beta[0])

Y_pred=x@beta

plt.figure(figsize=(10, 6))
plt.scatter(x[:, 1], y, color='blue', label='Actual Data')  # Plot actual data points
plt.plot(x[:, 1], Y_pred, color='red', label='Regression Line')  # Plot regression line
plt.title('Multiple Linear Regression using Matrix Method')
plt.xlabel('Feature 1')
plt.ylabel('Target Variable')
plt.legend()
plt.grid()
plt.show()
```
