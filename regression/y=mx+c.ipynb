```python
x_list = [1,2,3,4,5]
y_list = [1.2,1.8,2.6,3.2,3.8]

x_count = len(x_list)
y_count = len(y_list)

x_bar = 0
y_bar = 0

for x,y in zip(x_list,y_list):
    x_bar += x
    y_bar += y

x_bar = x_bar / x_count
y_bar = y_bar / y_count
print(x_bar)
print(y_bar)

y_y_bar = []
x_x_bar = []

for x,y in zip(x_list,y_list):
    x_x_bar.append(x - x_bar)
    y_y_bar.append(y - y_bar)

print(x_x_bar)
print(y_y_bar)

x_x_bar_sqr = []
for x in x_x_bar:
    x_x_bar_sqr.append(x*x)

print(x_x_bar_sqr)

numerator = []
den = 0
for x,y in zip(x_x_bar, y_y_bar):
    numerator.append(x * y)

for x in x_x_bar_sqr:
    den += x

print(numerator)
num = 0
for x in numerator:
    num += x

m = num/den
print(num)
print(den)
print("Slope = "+str(m))

c = y_bar - (m * x_bar)
print("Constant = "+str(c))

new_y = []
for y in x_list:
    new_y.append((m*x) + c)

print("New Value of y = ")
print(new_y)

error_rate = []
for y,y1 in zip(y_list,new_y):
    error_rate.append((y1 - y) * (y1 - y))
print("Error Rate of y = ")
print(error_rate)
```
